```python
images_path = []
labels = []
# root = os.path.join(path, 'raw-img')
root = '../task2/data/animals10/raw-img'

# for each animal (separate folder for each)
for label in os.listdir(root):
    # join the animal folder to path
    path = os.path.join(root, label)
    # add first 1460 images for each animal
    for img in os.listdir(path)[:1000]:
        # add image path
        images_path.append(os.path.join(path, img))
        # add label (animal name)
        labels.append(label)
```

```python
data = pd.DataFrame({
    'images_path' : images_path,
    'labels' : labels
})
data.head()
```

```python
label = {"cane": "dog",
         "cavallo": "horse",
         "elefante": "elephant",
         "farfalla": "butterfly",
         "gallina": "chicken",
         "gatto": "cat",
         "mucca": "cow",
         "pecora": "sheep",
         "ragno" : "spider",
         "scoiattolo": "squirrel"}

label_to_index = {key: idx for idx, key in enumerate(label.keys())}
index_to_label = {idx: value for key, value in label.items() for idx, k in enumerate(label.keys()) if k == key}
```

```python
data['labels_num'] = data['labels'].map(label_to_index)
data['labels'] = data['labels'].map(label)
data.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data,x='labels',alpha = 0.5)
plt.tight_layout()
```

```python
fig,ax = plt.subplots(2, 8, figsize=(12, 4))
ax = ax.flatten()

for idx in range(len(ax)):
    rd_id = np.random.randint(0,data.shape[0])
    image = plt.imread(data['images_path'].iloc[rd_id])
    ax[idx].imshow(image)
    ax[idx].set_title(data['labels'].iloc[rd_id], fontsize = 10)
    ax[idx].axis('off')

plt.tight_layout()
plt.show()
```

```python
all_images = []
def processing_image(image_path, size=(224,224)):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, size)
    return image

for i in data['images_path'].values:
    image = processing_image(i)
    all_images.append(image)
```

```python
def show_random_images(images, n=5):
    random_images = random.sample(images, n)
    plt.figure(figsize=(15, 5))

    for i, img in enumerate(random_images):
        plt.subplot(1, n, i + 1)
        plt.imshow(img)
        plt.axis("off")

    plt.show()

# Show 5 random images (change n to any number)
show_random_images(all_images, n=10)
```

```python
all_images = np.array(all_images)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    all_images,data['labels_num'],
    test_size = 0.1,
    random_state=42,
    stratify=data['labels_num']
)

x_train, x_valid, y_train, y_valid = train_test_split(
    x_train,y_train,
    test_size=0.1,
    random_state=42,
    stratify=y_train
)

print(x_train.shape,y_train.shape)
print(x_valid.shape, y_valid.shape)
print(x_test.shape, y_test.shape)
```

```python
aug = ImageDataGenerator(
    rescale=1./255,
    rotation_range = 0.18,
    zoom_range = 0.15,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip = True,
    fill_mode = 'nearest'
)
```

```python
vgg16 = VGG16(
    input_shape = (224,224,3),
    include_top = False,
)
```

```python
for layer in vgg16.layers:
    layer.trainable = False
```

```python
x = vgg16.layers[-1].output
x = Flatten()(x)
x = Dense(512)(x)
x = Activation('relu')(x)
output = Dense(10, activation='softmax')(x)
vgg16_1 = Model(vgg16.input,output)
vgg16_1.summary()
```

```python
learning_rate = 0.001
epochs = 50
batch_size = 32
factor = 0.2

vgg16_1.compile(
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy'],
    optimizer = keras.optimizers.Adam(learning_rate = learning_rate),
)
```

```python
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor= factor, patience=3, min_lr=1e-7)
checkpoint_cb = ModelCheckpoint('../task2/vgg16/model_epoch_{epoch:02d}.keras', save_freq='epoch')
```

```python
history = vgg16_1.fit(
    aug.flow(x_train, y_train, batch_size=batch_size),
    validation_data = (x_valid,y_valid),
    epochs = epochs,
    callbacks = [early_stopping , reduce_lr, checkpoint_cb]
)
```

```python
vgg16_1.save('../task2/vgg16/vgg16_model.keras')
```

```python
vgg16_load = keras.models.load_model('../task2/vgg16/vgg16_model.keras')
```
