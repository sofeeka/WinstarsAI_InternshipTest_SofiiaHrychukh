```python
import numpy as np
import matplotlib.pyplot as plt

from task2.inference import inference
from task2.utils.translate import index_to_label
```

## Examples
`animal_path` containing image path of an image of the animal
`animal_text` containing text with the animal

```python
cat_img = '../task2/data/test_examples/cat.jpg'
dog_img = '../task2/data/test_examples/dog.jpeg'
cow_img = '../task2/data/test_examples/cow.jpeg'
horse_img = '../task2/data/test_examples/horse.jpeg'
sheep_img = '../task2/data/test_examples/sheep.jpeg'
spider_img = '../task2/data/test_examples/spider.jpeg'
chicken_img = '../task2/data/test_examples/chicken.jpeg'
elephant_img = '../task2/data/test_examples/elephant.jpeg'
squirrel_img = '../task2/data/test_examples/squirrel.jpeg'
butterfly_img = '../task2/data/test_examples/butterfly.jpeg'
```

```python
cat_text = 'Oh this must be Moomin, my favourite cat.'
dog_text = 'There is a dog in the picture.'
cow_text = 'I can see a big black and white cow.'
horse_text = 'I see a horse here.'
sheep_text = 'Are those sheep in the picture?'
spider_text = 'The spider on the picture is scary.'
chicken_text = 'I see a chicken with a butterfly.'
elephant_text = 'An elephant is in the picture.'
squirrel_text = 'I can see a squirrel eating in the picture.'
butterfly_text = 'I can see a butterfly.'
```

```python
def make_prediction(text, img_path):
    truth_val, doc, predictions = inference(text, img_path)
    print(f'text: {text}')
    print(f'extracted: {doc.ents}')
    print(f'predictions: {predictions}')
    print(f'predicted label: {index_to_label[np.argmax(predictions)]}')
    print(f'\ntruth_val: {truth_val}')
    
    image = plt.imread(img_path)
    plt.imshow(image)
    plt.axis("off")
    plt.show()
```

## Positive answers

```python
make_prediction(cat_text, cat_img)
```

```python
make_prediction(dog_text, dog_img)
```
